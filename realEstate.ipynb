```python
import requests
from bs4 import BeautifulSoup


base_url ="https://pythonizing.github.io/data/real-estate/rock-springs-wy/LCWYROCKSPRINGS/t=0&s="




scrapList = []

# Loop through each webpage
'''page number could change so before executing check 
    the webpage for how the page number is given '''

for page in range(0,int(page_nr)*10,10):
    req= requests.get(base_url+str(page)+".html")
    content= req.content
    soup= BeautifulSoup(content, "html.parser")
    all = soup.find_all('div', {'class':'propertyRow'})

    # Loop through each listing for informations
    for item in all:
        tempDict = {}

        tempDict["Price"]= item.find("h4",{"class":"propPrice"}).text.replace("\n","").replace(" ","")
        tempDict["Address"]= item.find_all("span", {"class", "propAddressCollapse"})[0].text
        try:
            tempDict["Locality"]= item.find_all("span", {"class", "propAddressCollapse"})[1].text
        except:
            tempDict["Locality"]= "NIL"
        try:
            tempDict["Sqft"] = item.find_all("span", {"class", "infoSqFt"})[0].find("b").text
        except:
            tempDict["Sqft"]= "NIL"
        try:
            tempDict["Beds"] = item.find_all("span", {"class", "infoBed"})[0].find("b").text
        except:
            tempDict["Beds"]= "NIL"
        try:
            tempDict["FullBath"] = item.find_all("span", {"class", "infoValueFullBath"})[0].find("b").text
        except:
            tempDict["FullBath"]= "NIL"
        try:
            tempDict["HalfBath"]= item.find_all("span", {"class", "infoValueHalfBath"})[0].find("b").text
        except:
            tempDict["HalfBath"]= "NIL"
        # to get multiple values that is inside a div(span groups)
        for column_group in item.find_all("div",{"class":"columnGroup"}):
            for feature_group, feature_name in zip(column_group.find_all("span", {"class": "featureGroup"}), column_group.find_all("span", {"class": "featureName"})):
                if "Lot Size" in feature_group.text:
                    tempDict["LotSize"] = feature_name.text
                    
        scrapList.append(tempDict)
```

```python
import pandas
df= pandas.DataFrame(scrapList)
df
```

```python
df.to_csv("scappedFiles.csv")
```
